## Energy vs time

To produce the energy-time graphs (___Fig. 2___ and ___Fig. 3___) I have used two different sets of equations (___eq.___ $\eqref{eq:energy1}$ and ___eq.___ $\eqref{eq:energy2}$). Both capture the failed behavior of the Euler method and the relatively correct behavior of the Euler-Cromer method. The difference between the two graphs/equations is the way they show the potential energy of the pendulum. The problem with ___eq.___ $\eqref{eq:energy2}$ is that there is no $1 - \cos\theta$. So for example if we released the pendulum from $\theta = 90^{\circ}$ with initial velocity of zero then the kinetic energy would be zero because the initial velocity is zero and the potential energy would be zero because $\cos(90)$ is equal to zero. So, the total energy of the pendulum would be zero which does not make sense! However, we know that at $\theta = 90^{\circ}$ pendulum must have its maximum potential energy since it is located at a height equal to the length of the pendulum and we can get this maximum value if we use ___eq.___ $\eqref{eq:energy1}$. Besides, in ___eq.___ $\eqref{eq:energy2}$ potential energy is subtracted from the kinetic energy of the system! This subtraction may be explained by arguing that the system is doing work on the environment and the system loses energy as a result of that but I am not sure how valid this argument would be. But the main problem remains that why do we not see the $1 - \cos\theta$ term in ___eq.___ $\eqref{eq:energy2}$?

The reader can see the plots generated using ___eq.___ $\eqref{eq:energy1}$ in [___Fig. 2___](#Fig.-2-Energy-vs.-Time)


\begin{equation}
E = \frac{1}{2} m  L^2 \omega^2 + m g L (1 - \cos\theta) 
\label{eq:energy1} \tag{1}
\end{equation}

And for seeing the graphs generated by ___eq.___ $\eqref{eq:energy2}$ (below) see [___Fig. 3___](#Fig.-3-Energy-vs.-Time).

\begin{equation}
E = \frac{1}{2} m  L^2 \omega^2 - m g L \cos\theta
\label{eq:energy2} \tag{2}
\end{equation}

The above set of plots (x and y components of position vectors) are the result of the three different runs with ___dt___ equal to $0.0001$, $0.001$, and $0.01$ for a total run of ___t = 5___. As stated above object ___two___ has twice the mass of object ___one___. Also, object ___one___ has an initial velocity of $V_0 = 1 \frac{m}{s}$ in the direction of x axis! For the analysis of these plots one must consider two important points. First, these simulations are not based on an ___adaptive___ Runge-Kutta method. This means, the simulations are based on fixed stepsizes (i.e. $dt$). Since all simulations are run for a period of 5 seconds (the unit "___second___" is chosen arbitrarily) with a fixed stepsize the simulation with a smaller stepsize will have fewer number of time steps to complete the simulation (it will go faster). However, the error accumulation will be larger for simulations with smaller larger stepsized (fewer number of steps). This analysis is an attempt to illustrate this trade-off between simulation run-time and accuracy of the results! Second, one must pay speciall attention to the collision moment(s) in these simulations. As the to objects approach one another the change in their attraction force changes rapidly. A similar behavior is seen in the oribt of planets around the sun. When planets approach the sun they move faster according to the 
Kepler's laws of planetary motion. In this scenario a small timestep (___dt___) will capture the behavior of the two objects better than a bigger timestep. In other words, the smaller the time step the better behvior prediction/simulation. This is, specially, the case for the position plots above. As it can be seen on the left plot (the simulation with the smallest timestep $dt = 0.0001$) the behvior of the two obejcts has been captured better than the other two runs, considerably.

There seems to be a similar pattern in all plots and that is the spikes in velocity where these spikes correspond to the the moments where the objects come in close distance to one another. At the bottom plot we can only observe one spike which is expected since the two objects meet each other only once. The oscillatory pattern of the middle plot is also expected as the same pattern was observed in the distance plot of the simulation with $dt = 0.001$ above. While the three plots have the spikes in common (though in different numbers and frequency) only the middle and bottom plot show two nearly straight lines. These straight lines happen after the two object experience their closest distance after which the two objects move away (see distance plots above). The intersting point is that in the middle plot there is a sudden increase in the velocity which indicates the moment of final ecounter. After that the objects move away rapidly at each time step. The rapid increase in the distance between the two objects causes a sharp decrease in the magnitude of the force they exert on each other which results in turn results in a sharp decrease in change in acceleration and, accordingly, in velocity of the two objects. As a result the two object move away with realitvely constant velocities. However, unlike the middle plot, the bottom plot shows a sudden decrease in velocity before the two objects move away from each other. A possible explanation for this different behavior could be the shortest distance between the two objects at their final encounter. The minimum distance between the two objects in the simulation with $dt = 0.001$ (middle plot) is about $\approx$ $0.019134$ while the minimumn distance for the simulation with $dt. 0.01$ (bottom plot) is about $\approx$ $0.085987$. (see the paragraph above on the acquisition of the minimum distance and the time at which the minimum distance occured.) As it can be seen the minimum distance in simulation with $dt = 0.001$ is about $\approx$ $4.5$ smaller than the minimum distance in simulation with $dt = 0.01$. This means that the objects in simulation with $dt = 0.001$, in their final encounter, experience a distance about $\approx$ $4.5$ smaller than the distance the two objects experience in their final encounter in the simulation with $dt = 0.01$. It is worth noting that the garvitational force between the two objects is proportional to the distance between the two objects cubed ($F$ $\propto$ $\frac{1}{|r_{12}|^3}$). This means that at the simulation with $dt = 0.001$where the minimum distance between the two objects is 4.5 times greater than the minimum distance in the simulation with $dt = 0.01$ will result in a force about $(4.5)^3$ $\approx$ $91$ times greater than the force the two bodies exert on each other in the simulation with $dt = 0.01$.

Note that for making movies of the three-body problem you need to activate and comment out the right input file. There are two input files below one of which has to be commented out. Only read the ___threeBodyRunge4Adaptive.csv___ or ___threeBodyRunge4.csv___.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# https://community.dur.ac.uk/joshua.borrow/blog/posts/making_research_movies_in_python/

runge4_001 = pandas.read_csv("threeBodyRunge4Adaptive.csv", header = None, names = ['time', 'x1', 'y1', 'x2', 'y2', 'x3', 'y3', 'vx1', 'vy1', 'vx2', 'vy2', 'vx3', 'vy3'])

#runge4_001 = pandas.read_csv("threeBodyRunge4.csv", header = None, names = ['time', 'x1', 'y1', 'x2', 'y2', 'x3', 'y3', 'vx1', 'vy1', 'vx2', 'vy2', 'vx3', 'vy3'])

time_001 = runge4_001.time.to_numpy()

all_x1_data   = runge4_001.x1.to_numpy()
all_y1_data   = runge4_001.y1.to_numpy()

all_x2_data   = runge4_001.x2.to_numpy()
all_y2_data   = runge4_001.y2.to_numpy()

all_x3_data   = runge4_001.x3.to_numpy()
all_y3_data   = runge4_001.y3.to_numpy()

def animate(frame, coeff, line1, line2, line3, line4, line5, line6):
#     global all_x_data, all_y_data
    
    current_x1_data = all_x1_data[: coeff * frame + 1]
    current_y1_data = all_y1_data[: coeff * frame + 1]
    
    current_x2_data = all_x2_data[: coeff * frame + 1]
    current_y2_data = all_y2_data[: coeff * frame + 1]
    
    current_x3_data = all_x3_data[: coeff * frame + 1]
    current_y3_data = all_y3_data[: coeff * frame + 1]
    
    line1.set_xdata(current_x1_data)
    line1.set_ydata(current_y1_data)
    line1.set_color('red')
    
    line2.set_xdata(current_x2_data)
    line2.set_ydata(current_y2_data)
    line2.set_color('green')
    
    line3.set_xdata(current_x3_data)
    line3.set_ydata(current_y3_data)
    line3.set_color('blue')

    current_x1_data = all_x1_data[coeff * frame - 1: coeff * frame]
    current_y1_data = all_y1_data[coeff * frame - 1: coeff * frame]
    
    current_x2_data = all_x2_data[coeff * frame - 1: coeff * frame]
    current_y2_data = all_y2_data[coeff * frame - 1: coeff * frame]
    
    current_x3_data = all_x3_data[coeff * frame - 1: coeff * frame]
    current_y3_data = all_y3_data[coeff * frame - 1: coeff * frame]
    
    line4.set_xdata(current_x1_data)
    line4.set_ydata(current_y1_data)
    line4.set_color('red')
    line4.set_marker('o')
    
    line5.set_xdata(current_x2_data)
    line5.set_ydata(current_y2_data)
    line5.set_color('green')
    line5.set_marker('o')
    
    line6.set_xdata(current_x3_data)
    line6.set_ydata(current_y3_data)
    line6.set_color('blue')
    line6.set_marker('o')
    
    time_elapsed = time_001[coeff * frame: coeff * frame + 1]
    time_text.set_text('time = %.1f' % time_elapsed)
    
    return (line1, line2, line3, line4, line5, line6)

# Now we can do the plotting!
fig, ax = plt.subplots(1, figsize = (8,8))
# Initialise our line
line1, = ax.plot([0], [0])
line2, = ax.plot([0], [0])
line3, = ax.plot([0], [0])

line4, = ax.plot([0], [0])
line5, = ax.plot([0], [0])
line6, = ax.plot([0], [0])
time_text = ax.text(0.05, 0.9, '', transform=ax.transAxes, fontsize=13)
# Have to set these otherwise we will get one ugly plot!
ax.set_xlim(-5, 5)
ax.set_ylim(-5, 7)

ax.set_title(r'Three-body problem (4th order Runge-Kutta)', fontsize=15)
ax.set_xlabel(r'X coordinate', fontsize=13)
ax.set_ylabel(r'Y coordinate', fontsize=13)

# Make me pretty
fig.tight_layout()

coeff = 1
animation = FuncAnimation(
    # Your Matplotlib Figure object
    fig,
    # The function that does the updating of the Figure
    animate,
    # Frame information (here just frame number)
    np.arange(546),
    # Extra arguments to the animate function
    fargs=[coeff, line1, line2, line3, line4, line5, line6],
    # Frame-time in ms; i.e. for a given frame-rate x, 1000/x
    interval=1000 / 20
)
animation.save("exersice2.mp4")
```
