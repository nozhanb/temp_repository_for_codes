```python
def mcmc_test(initial_value, cov_matrix):
    
    old_x_y      = initial_value
    theta_list_x = []
    theta_list_y = []
    count        = 0
    while count < 10000:
    
        # proposal (returns a value that is not probability). Must be symmetric!
        pr_x = norm.rvs(old_x_y[0], 1)
        pr_y = norm.rvs(old_x_y[1], 1)
        current_x_y = numpy.array([pr_x, pr_y])
        
        # Uniform distribution (returns a value that is not probability)
        uni  = uniform.rvs()

        # acceptance
        numerator   = multivariate_normal([0, 0], cov = cov_matrix)
        denominator = multivariate_normal([0, 0], cov = cov_matrix)
        ratio       = min(1, numerator.pdf(current_x_y)/denominator.pdf(old_x_y))

        if ratio > uni:
            old_x_y = current_x_y
        if ratio <= uni:
            old_x_y = old_x_y
        
        theta_list_x.append(old_x_y[0])
        theta_list_y.append(old_x_y[1])

        count += 1
        
#     return numpy.array([theta_list_x, theta_list_y]).reshape(count, current_x_y.size)

    return numpy.array(theta_list_x), numpy.array(theta_list_y)

    
```

```python
def log_target_distribution(x_i, y_i, var_y_i, param_array):
    
    m   = param_array[0]
    b   = param_array[1]
    p_b = param_array[2]
    y_b = param_array[3]
    v_b = param_array[4]
    print("This is p_b ---> ", p_b)
    if p_b <= 0.0:
        ln_prior = numpy.NINF
        
        ln_likeli = numpy.NINF
        
    elif p_b > 1:
        ln_prior = numpy.NINF
        
        ln_likeli = numpy.NINF
    
    else:
        
        # Bernoulli distribution
        bernoulli_array = bernoulli.rvs(p_b, size = 20)
        number_of_ones  = (bernoulli_array > 0.0).sum()

        # Prior distribution
        binomial_distribution = binom.pmf(k = number_of_ones, n = 20, p = (1 - p_b))
        prior_m        = uniform.pdf(m,   0, 10)
        prior_b        = uniform.pdf(b,   -100, 200)
        prior_p_b      = uniform.pdf(p_b, 0, 1)
        prior_y_b      = uniform.pdf(y_b, 300, 200)
        prior_v_b      = uniform.pdf(v_b, 80, 20)

        print("probability ---> ", binomial_distribution, prior_m, prior_b, prior_p_b, prior_y_b, prior_v_b)
        print("parameters ---> ", number_of_ones, m, b, p_b, y_b, v_b)

    #     prior_array    = numpy.array([binomial_distribution, prior_m, prior_b, prior_p_b, prior_y_b, prior_v_b])

        prior        = binomial_distribution * prior_m * prior_b * prior_p_b * prior_y_b * prior_v_b
        ln_prior     = numpy.log(prior)
    
        # Likelihood distribution
        linear_term  = ((1 - p_b)/numpy.sqrt(6.28*var_y_i))*numpy.exp(-((y_i - m*x_i - b)**2)/(2*var_y_i))

        outlier_term = ((p_b)/numpy.sqrt(6.28*(v_b+var_y_i)))*numpy.exp(-((y_i - y_b)**2)/(2*(v_b+var_y_i)))

        likelihood   = numpy.prod(linear_term + outlier_term)
    
        ln_likeli    = numpy.log(likelihood)
        
        print(ln_prior, prior)
        
    # Final output!!
    output           = ln_likeli + ln_prior
    
    return output
```

```python
def mcmc_test(x_i, y_i, var_y, old_proposal_array):
    
    theta_list_x = []
    theta_list_y = []
    count        = 0
    while count < 100:

        # proposal (returns a value that is not probability). Must be symmetric!
        
        m   = norm.rvs(old_proposal_array[0], 1)
        b   = norm.rvs(old_proposal_array[1], 1)
        p_b = norm.rvs(old_proposal_array[2], 1)
        y_b = norm.rvs(old_proposal_array[3], 1)
        v_b = norm.rvs(old_proposal_array[4], 1)
        
        new_proposal_array = numpy.array([m, b, p_b, y_b, v_b])
        
#         print("old proposal ---> ", old_proposal_array)
#         print("new proposal ---> ", new_proposal_array)
        # Uniform distribution (returns a value that is not probability)s
        uni  = uniform.rvs()

        # acceptance
        numerator    = log_target_distribution(x_i, y_i, var_y, new_proposal_array)
        denominator  = log_target_distribution(x_i, y_i, var_y, old_proposal_array)
        subtraction  = min(1, numerator - denominator)
            
        if subtraction > numpy.log(uni):
            old_proposal_array = new_proposal_array
            theta_list_y.append(1)
        if subtraction <= numpy.log(uni):
            old_proposal_array = old_proposal_array
        
        print(" subtraction ---> ", numerator, denominator, subtraction, numpy.log(uni))
        theta_list_x.append(old_proposal_array)

        count += 1
        
#     return numpy.array([theta_list_x, theta_list_y]).reshape(count, current_x_y.size)

    return numpy.array(theta_list_x), numpy.array(theta_list_y)

    
```

```python
def climate_target(param_array, t, y_obs, y_err):
    
    # Parameter assignment
    a_1    = param_array[0]
    a_2    = param_array[1]
    a_3    = param_array[2]
    p_1    = param_array[3]
    p_2    = param_array[4]
    p_3    = param_array[5]
    t_null = param_array[6]
    
    # ln_prior evaluation
    if 0.0 < a_1 < 5.0 and 0.0 < a_2 < 5.0 and 0.0 < a_3 < 5.0 and 10000. < p_1 < 200000 and 10000. \
        < p_2 < 200000 and 10000. < p_3 < 200000 and -10.0 < t_null < 0:
    
        ln_prior = 0.0
        
    else:
        ln_prior = -numpy.inf
    
    # ln_likelihood evaluation
    term_one   = a_1 * numpy.sin((6.28 * t)/p_1)
    term_two   = a_2 * numpy.sin((6.28 * t)/p_2)
    term_three = a_3 * numpy.sin((6.28 * t)/p_3)
    
    y_model    = term_one + term_one + term_one + t_null

    ln_likeli  = -0.5 * (((y_obs - y_model)/y_err)**2).sum()
        
    return ln_prior + ln_likeli
```

```python
def model(theta, age = age_array):
    a1,a2,a3,p1,p2,p3,T0 = theta
    return a1*numpy.sin(2*numpy.pi*age/p1) + a2*numpy.sin(2*numpy.pi*age/p2) + a3*numpy.sin(2*numpy.pi*age/p3) + T0
```

```python
data     = (age_array, temp_array, temp_sigma)
nwalkers = 128
niter    = 5000
initial  = numpy.array([1.0, 1.0, 1.0, 26000., 41000., 100000., -4.5])
ndim     = len(initial)
p0       = [numpy.array(initial) + 1e-7 * numpy.random.randn(ndim) for i in range(nwalkers)]
```

```python
def main(p0, nwalkers, niter, ndim, lnprob, data):
    sampler  = emcee.EnsembleSampler(nwalkers, ndim, climate_target, args=data)

    print("Running burn-in...")
    p0, _, _ = sampler.run_mcmc(p0, 100)
    sampler.reset()

    print("Running production...")
    pos, prob, state = sampler.run_mcmc(p0, niter)

    return sampler, pos, prob, state
```

```python
sampler, pos, prob, state = main(p0,nwalkers,niter,ndim,climate_target,data)
```
